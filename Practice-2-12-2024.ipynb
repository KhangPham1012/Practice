# For loop
Trong cell này chúng ta sẽ sử dụng vòng lặp for

```python
My_hobbies = {"Books", "Music"}
for hobby in My_hobbies:
    print("-" + hobby)
```

# Average 
Trong cell này chúng ta sẽ tính giá trị trung bình cho một dãy số:

```python
def calculate_average(numbers):
   
    if len(numbers) == 0:
        return 0 
    total = sum(numbers) 
    average = total / len(numbers) 
    return average


numbers = [20, 20, 90, 500, 700]
average = calculate_average(numbers)
print("Giá trị trung bình của danh sách là:", average)
```
